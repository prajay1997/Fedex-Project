### **Columns In our Datasets**



*   Year: The Year the data was collected
*    Month: The Month in which the data was collected 
*   DayofWeek: The day of Week
*   Actual_Shipment_Time: The Actual time when the package was sent for   shipment. (ex: 1955 means 19 hours and 55 minutes i.e 7:55 PM) 
*  Planned_Shipment_Time: The time when the package should have been sent for shipment. (ex: 1955 means 19 hours and 55 minutes i.e 7:55 PM) 
*   Planned_Delivery_Time: The time when the package should be delivered. (ex: 1955 means 19 hours and 55 minutes i.e 7:55 PM)
*   Carrier_Name: The name of the Carrier which carried the package.
*   Carrier_Num: The number of the Carrier which carried the package.
*   Planned_TimeofTravel: The estimated time to reach from Source to Destination. ( in minutes) 
*   Shipment_Delay: The time by which the package was shipped late. (in minutes. Negative value indicates that the package was shipped early. Ex: 4 indicates that the package was shipped 4 minutes late, whereas, -4 indicates that the package was shipped 4 minutes early)
*  Source: The place from which the package was shipped.
*  Destination: The place at which the package was delivered.
*  Distance: Distance between Source and Destination in miles.
*  Delivery_Status: Whether it got delivered at right time or not. (Dependent Variable)

```python
import pandas as pd
import numpy as np
import matplotlib.pylab as plt
from sklearn.cluster import	KMeans
```

```python
# Generating random uniform numbers 
X = np.random.uniform(0,1,50)
Y = np.random.uniform(0,1,50)
df_xy = pd.DataFrame(columns=["X","Y"]) #  Creating the dataframe
df_xy.X = X
df_xy.Y = Y
```

```python
model1 = KMeans(n_clusters = 3).fit(df_xy)
df_xy.plot(x = "X", y = "Y", c = model1.labels_, kind="scatter", s = 10, cmap = plt.cm.coolwarm)
```

```python
#  Kmeans on fedex Data set 
fed = data1
fed.head()
```

```python
fed.dtypes
```

```python
fed.shape
```

```python
# Taking the selected variables for the clustring
fed1 = fed.iloc[:,[15,4,6,8,10,11,14]]
```

```python
fed1.head()
```

```python
# Normalization function 
def norm_func(i):
    x = (i - i.min())	/ (i.max() - i.min())
    return (x)
```

```python
# Normalized data frame (considering the numerical part of data)
fed_norm = norm_func(fed1.iloc[:,1:])
fed_norm.head()
```

```python
###### scree plot or elbow curve ############
TWSS = []
k = list(range(1,5))
```

```python
for i in k:
  kmeans = KMeans(n_clusters = i)
  kmeans.fit(fed_norm)
  TWSS.append(kmeans.inertia_)
TWSS
```

```python
# Scree plot 
plt.plot(k, TWSS, 'ro-')
plt.xlabel("No_of_Clusters")
plt.ylabel("total_within_SS")
```

```python
# Selecting 2 clusters from the above scree plot which is the optimum number of clusters 
model = KMeans(n_clusters = 2)
model.fit(fed_norm)
```

```python
model.labels_                  # getting the labels of clusters assigned to each row 
mb = pd.Series(model.labels_)  # converting numpy array into pandas series object 
fed1['clust'] = mb              # creating a  new column and assigning it to new column 
fed1.head()
```

```python
mean_ship_delay = fed1.iloc[:, 1 :].groupby(fed1.clust).mean()
mean_ship_delay.transpose()
```

From the above table we can say that as the mean shipment delay time is 1.68 min and if the planned / actual shipment will do at 9.41 or 9.51 am and the planned delivery time will be 11.27 am then and the averahe distnce is high then the delevery will be on time.

  and 

if the mean shipment delay time is 9 min and the  mean or the actual/planned shipment time is 17.25/ 17.12 at evening and if average distance is 659 km then the delivery will be late.

 #Now we will find out the most busiest place and its networks to the other places for the deliveries by using the network analysis.





```python
import networkx as nx
```

```python
g = nx.Graph()                                                                 # craeting a empty graph with no. nodes and no. edges

g = nx.from_pandas_edgelist(data1, source = 'Source', target = "Destination")    # takes source and destination nodes from the dataframe 

print(nx.info(g))                                                              # it gives the information about no of edges, nodes and average degree.
```

```python
# degree centrality
d = nx.degree_centrality(g)
pd.Series(d).sort_values(ascending = False).head(10)
```

we can see ATL, ORD, DFW, DEN,MSP,DTW , IAH, SLC,CVG, LAS are the top 10  places (Airports) which has the highest degree of centrality.

```python
plt.figure(figsize = (20,10))
pos = nx.spring_layout(g, k= 0.2)
nx.draw_networkx(g,pos, node_size = 10, node_color = 'blue') 
```

```python
# closeness Centrality
c = nx.closeness_centrality(g)
pd.Series(c).sort_values(ascending = False).head(10)
```

Again ATL, ORD, DFW, DEN,MSP,DTW , IAH, SLC,CVG, LAS has the highest closeness centrality.


```python
# Betweenness Centrality
b = nx.betweenness_centrality(g)
pd.Series(b).sort_values(ascending = False).head(10)
```

ATL, DFW, SLC, MSP, ORD, DEN, IAH, DTW, ANC, SEA are the top 10 airports which has the  highest betweenness centrality. Out of this ATL, ORD, DFW, DEN, MSP, DTW , IAH, SLC this airports has the highest degree of centrality and closeness centrality also.


```python
# Eigen vector Centrality 
e = nx.eigenvector_centrality(g)
pd.Series(e).sort_values(ascending = False).head(10)
```

Again ATL, ORD, DFW, DEN, MSP, DTW , IAH, SLC are the top 10 airports which has highest eigen vector centrality.means this are the airports which having the conectivity with the other more important airports.

```python
# Cluster cofficient
clcoeff = nx.clustering(g)
pd.Series(clcoeff).sort_values(ascending = False).head(10)
```

DRO, GTF,BTR, MDT, COD etcare the airports has the highest clustering coefficient.

```python
print(clcoeff['ATL'],clcoeff['ORD'], clcoeff['DFW'], clcoeff['DEN'], clcoeff['MSP'], clcoeff['DTW'], clcoeff['IAH'], clcoeff['SLC'])
```

we can see the cluster coefficients for the airports ATL, ORD, DFW, DEN, MSP, DTW , IAH, SLC are  not much bad.


```python
# average clustering
ac = nx.average_clustering(g)
print(ac)
```
